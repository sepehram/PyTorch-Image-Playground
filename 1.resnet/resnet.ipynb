```python
from torchvision import models
dir(models)
```

```python
alexnet = models.AlexNet()
```

```python
resnet = models.resnet101(pretrained=True)
```

```python
resnet
```

```python
from torchvision import transforms
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean = [0.485, 0.456, 0.406],
        std = [0.229, 0.224, 0.225]
    )])
```

```python
from PIL import Image
img = Image.open("./candy.jpeg")
img.show()
```

```python
img_t = preprocess(img)
img_t
```

```python
import torch
batch_t = torch.unsqueeze(img_t, 0)
batch_t
```

```python
resnet.eval() #put in eval mode!
out = resnet(batch_t)
with open('./imagenet_classes.txt') as f:
    labels = [line.strip() for line in f.readlines()]
```

```python
_ ,index = torch.max(out,1)
percentage = torch.nn.functional.softmax(out,dim=1)[0] * 100
labels[index[0]], percentage[index[0]].item()
```

```python
_, indices = torch.sort(out, descending=True)
[(labels[idx], percentage[idx].item()) for idx in indices[0][:5]]
```

```python
img2 = Image.open("candy2.jpeg")
img2_t = preprocess(img2)
batch2_t = torch.unsqueeze(img2_t, 0)
resnet.eval()
out2 = resnet(batch2_t)
_, index2 = torch.max(out2,1)
percentage2 = torch.nn.functional.softmax(out2,dim=1)[0] * 100
labels[index2[0]], percentage2[index2[0]].item()
```

```python
_, indices2 = torch.sort(out2, descending=True)
[(labels[idx], percentage2[idx].item()) for idx in indices2[0][:10]]
```
